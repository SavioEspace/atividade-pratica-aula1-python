## 5) Atividade 3 — KPIs do negócio

Considere que você é analista de um varejista e precisa apresentar indicadores para a gestão.

Calcule os seguintes KPIs:

1. **Receita total**
2. **Lucro total**
3. **Ticket médio** (`Receita média por registro`)
4. **Quantidade total vendida**

Depois, escreva uma breve interpretação:
- O desempenho geral parece positivo?
- Receita e lucro caminham juntos?

```python
# KPIs principais
receita_total = df["Receita"].sum()
lucro_total = df["Lucro"].sum()
ticket_medio = df["Receita"].mean()
quantidade_total = df["Quantidade"].sum()

print(f"Receita total: R$ {receita_total:,.2f}")
print(f"Lucro total:   R$ {lucro_total:,.2f}")
print(f"Ticket médio:  R$ {ticket_medio:,.2f}")
print(f"Qtd. vendida:  {quantidade_total}")

# Interpretação
print("\nInterpretação:")
print("- O desempenho geral parece positivo, porque o lucro total é positivo.")
print("- Receita e lucro caminham juntos, mas não na mesma proporção, pois vender mais não garante margens maiores.")
```

## 6) Atividade 4 — Respondendo perguntas do gestor com dados

Use o dataset para responder:

1. **Qual canal gera mais receita?**
2. **Qual produto gera mais lucro?**
3. **Quais UFs concentram maior receita?**

Para cada resposta:
- mostre o código;
- apresente o resultado;
- escreva uma interpretação curta.

```python
# 1. Qual canal gera mais receita?
receita_por_canal = df.groupby("Canal_Venda", dropna=False)["Receita"].sum().sort_values(ascending=False)
print("Receita por canal:")
print(receita_por_canal)

print("\nCanal com maior receita:")
print(receita_por_canal.head(1))

# 2. Qual produto gera mais lucro?
lucro_por_produto = df.groupby("Produto")["Lucro"].sum().sort_values(ascending=False)
print("\nLucro por produto:")
print(lucro_por_produto)

print("\nProduto com maior lucro:")
print(lucro_por_produto.head(1))

# 3. Quais UFs concentram maior receita?
receita_por_uf = df.groupby("UF", dropna=False)["Receita"].sum().sort_values(ascending=False)
print("\nReceita por UF:")
print(receita_por_uf)

print("\nTop 5 UFs por receita:")
print(receita_por_uf.head(5))

print("\nInterpretação curta:")
print("- O canal líder em receita mostra onde a operação está mais forte.")
print("- O produto mais lucrativo indica melhor contribuição para resultado.")
print("- As UFs no topo concentram maior potencial comercial.")
```

## 7) Atividade 5 — Escolhendo a visualização certa

Sem necessariamente construir os gráficos ainda, responda:

a) Qual gráfico você usaria para mostrar a **evolução da receita ao longo do tempo**?  
b) Qual gráfico você usaria para **comparar a receita entre canais de venda**?  
c) Qual gráfico você usaria para mostrar a **distribuição dos preços unitários**?

Em seguida, gere **ao menos um gráfico** com `pandas` ou `matplotlib`.

```python
# Respostas teóricas:
print("a) Evolução da receita ao longo do tempo -> gráfico de linha")
print("b) Comparar receita entre canais -> gráfico de barras")
print("c) Distribuição dos preços unitários -> histograma")

# Gerando um gráfico de barras da receita por canal
import matplotlib.pyplot as plt

receita_por_canal = df.groupby("Canal_Venda", dropna=False)["Receita"].sum().sort_values(ascending=False)
receita_por_canal.plot(kind="bar", figsize=(8,5), title="Receita por Canal de Venda")
plt.ylabel("Receita")
plt.xlabel("Canal")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Desafio extra (opcional)

Faça uma pequena limpeza no dataset:

- trate valores nulos;
- padronize os nomes dos canais de venda;
- converta `Data_Venda` para formato de data.

Depois, compare os resultados **antes e depois da limpeza**.

```python
# Copiando o dataframe para limpeza
df_limpo = df.copy()

# Tratar valores nulos
df_limpo["Canal_Venda"] = df_limpo["Canal_Venda"].fillna("Não informado")
df_limpo["UF"] = df_limpo["UF"].fillna("Não informado")
df_limpo["Preco_Unitario"] = df_limpo["Preco_Unitario"].fillna(df_limpo["Preco_Unitario"].median())

# Padronizar nomes dos canais
df_limpo["Canal_Venda"] = df_limpo["Canal_Venda"].replace({
    "e-commerce": "E-commerce",
    "loja fisica": "Loja Física"
})

# Converter data
df_limpo["Data_Venda"] = pd.to_datetime(df_limpo["Data_Venda"])

print("Antes da limpeza:")
print(df.isnull().sum())
print("\nCategorias originais de Canal_Venda:")
print(df["Canal_Venda"].value_counts(dropna=False))

print("\nDepois da limpeza:")
print(df_limpo.isnull().sum())
print("\nCategorias padronizadas de Canal_Venda:")
print(df_limpo["Canal_Venda"].value_counts(dropna=False))

print("\nTipo da coluna Data_Venda após conversão:")
print(df_limpo["Data_Venda"].dtype)
```

Uma análise de dados não deve se limitar a fazer gráficos bonitos porque o objetivo real é gerar entendimento e apoiar decisões.
